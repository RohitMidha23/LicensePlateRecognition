```python
import pandas as pd 
```

```python
df = pd.read_json('/Users/rohit/LicensePlateRecognition/Indian_Number_plates.json',lines=True)
```

```python
x=[]
for i in range(237):
    x.append(df['content'][i])
x
```

```python
# WebScrapper to download the images 
import requests 
for i in range(237):
    image_url = x[i]
    r = requests.get(image_url)
    with open ('lpr'+str(i)+'.jpeg','wb') as f:
        f.write(r.content)
```
